# Day 1 - Python Basics

**Topics:**
- Variables & data types
- Numbers, strings, booleans
- Lists, tuples, dictionaries
- Basic operators
- Print and input functions
- Mini exercises

```python
# Numbers
x = 10      # integer
y = 3.5     # float

# Strings
name = "Dharma"

# Boolean
is_active = True

# Print variables
print(x, y, name, is_active)

# Check types
print(type(x), type(y), type(name), type(is_active))
```

```python
# List
fruits = ["apple", "banana", "cherry"]
print(fruits)
print(fruits[0])  # Access first element

# Tuple (immutable)
coords = (10, 20)
print(coords)

# Dictionary
person = {"name": "Dharma", "age": 22}
print(person)
print(person["name"])
```

```python
a = 10
b = 3

# Arithmetic
print(a + b, a - b, a * b, a / b, a % b, a ** b)

# Comparison
print(a > b, a == b, a != b)
```

```python
# Input from user
user_name = input("Enter your name: ")
print("Hello,", user_name)
```

```python
# Swapping of numbers without using third variable.
x = 5
y = 10
x = x + y
y = x - y
x = x - y
print(x,y)

# Printing all even numbers from 1 tp 20
for i in range(11):
    print(i*2, end = " ")
print("\n")

# Constructing dictionary with letters as keys and their AASCII as values
d = {"a": ord('a'),"b":ord('b'), "c":ord('c'), "d":ord('d'), "e":ord('e')}
print(d)
```
