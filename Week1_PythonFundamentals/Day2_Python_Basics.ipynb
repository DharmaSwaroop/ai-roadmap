```python
# Arithmetic Operations
a = 5
b = 3
print(a + b)
print(a - b)
print(a * b)
print(a/b) # Gives float value
print(a//b)
print(a%b)
print(a ** b) # This is a^b
```

```python
# Math Functions
import math
x = 2.9 
print(round(x)) # Rounds off the number
print(abs(-2.9))
print(math.ceil(x))
print(math.floor(x))
```

```python
# if, elif and else statements
is_hot = False
is_cold = True
if is_hot:
    print("It's a hot day")
    print("Dring more water")
elif is_cold:
    print("It's a cold day")
    print("Wear warm clothes")
else:
    print("Enjoy your day")    

# Exercise
price = 1000000
good_credit = True
if good_credit:
    down_payment = 0.1 * price
else:
    down_payment = 0.2 * price
print(f'The down payment is ${int(down_payment)}')
print("hello")
```

```python
# Logical Operators
has_high_income = False
has_good_credit = False
has_criminal_record = False

if has_high_income and has_good_credit:
    print("The applicant is applicable")
elif has_high_income or has_good_credit:
    print("The applicant is partially applicable")
elif has_good_credit or not has_criminal_record: # This wont be executed or checked if the previous elif is already true
    print("The applicant is eligible also")
else:
    print("The applicant is not applicable")
```

```python
# Comparison Operators

temperature = 30

if temperature > 30:
    print("It's a hot day")
else:
    print("It's not a hot day")

if temperature == 30:
    print("The temperature is 30")

# Exercise
name = "Dharma Swaroop"

if len(name) < 3:
    print("Name must be at least 3 characters")
elif len(name) > 50:
    print("The Max length of the name is 50 char")
else:
    print("Name looks good")
    print("The name is:", name)
```

```python
# While loops
i = 1
while i <= 5:
    print('*' * i)
    i = i + 1
print("Done")

# Exercise
secret_number = 9
guess_limit = 3
guess_count = 0
while guess_count < guess_limit:
    guess_count += 1
    guess = int(input('Guess:'))
    if guess == secret_number:
        print("You Won!")
        break
else:
    print("Sorry, you failed!")
```

```python
# for loops

for item in 'Python':
    print(item)
for item in ['Mosh', 'Dharma', 'Saroop']:
    print(item)
for item in range(10): # print 0 to 10
    print(item)
for item in range(5,7): # 7 wont be included
    print(item)

# Exercise

prices = [10,20,30,40,10]
total_price = 0
for price in prices:
    total_price += price
print("The total price is", total_price)
```

```python
# Nested loops
for x in range(4):
    for y in range(3):
        print(f'({x},{y})')

# Exercises
numbers = [5,2,5,2,2]
for count in numbers:
    output = ""
    for j in range(count):
        output += 'x'
    print(output)
```

```python
# Exercises

# sum of first n numbers using both for and while
n = int(input("Enter value for n:"))
sum1 = 0
for i in range(1,n+1):
    sum1 += i
print(f'The sum using for loop is {sum1}')

sum2 = 0
i = 1
while i <= n:
    sum2 += i
    i += 1
print(f'The sum using while loop is {sum2}')

# Count digits in a number:
number = int(input("Enter a number: "))
count_digits = 0
while number > 0:
    count_digits += 1
    number = number // 10
print(f'The number of digits are {count_digits}')

# Print triangle pattern with nested loops

for i in range(n):
    spaces = " "*(n-i-1)
    stars = ""
    for j in range(i*2+1):
        stars += '*'
    print(f'{spaces}{stars}')

# Reverse a string using a loop

string = 'DharmaSwaroop'
reverse = ''
for i in range(len(string)-1,-1, -1):
    reverse += string[i]
print(f'The reversed string of {string} is {reverse}')
```
