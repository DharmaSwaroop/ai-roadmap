```python
# Numbers
x = 10      # integer
y = 3.5     # float

# Strings
name = "Dharma"
course = '"Python for Beginners"'
course2 = course
course3 = '''
This is one more way of
declaring a string of multiple lines
'''

# Formated Strings
msg = f'{course} course is easy'

# Boolean
is_active = True

# Print variables
print(x, y, name, is_active)
print("The character name is " + name)
print("He is", is_active)
print("Printing the index of 'a' in name is", name.index("a"))
print("The course name is:", course)
print(course[0])
print(course[-1])
print(course[0:3])
print(course[0:])
print(course)
print(course3) # This will print empty lines in the front and end of the string
print(msg)

# String Methods
print(course.lower())
print(course.upper())
print(course.title())
print(len(course))
print(course.find('B'))
print(course.replace('Python','C++'))
print('Python' in course)

# Check types
print(type(x), type(y), type(name), type(is_active))
```

```python
# List
fruits = ["apple", "banana", "cherry"]
print(fruits)
print(fruits[0])  # Access first element

# Tuple (immutable)
coords = (10, 20)
print(coords)

# Dictionary
person = {"name": "Dharma", "age": 22}
print(person)
print("Person's name is", person["name"])
```

```python
a = 10
b = 3

# Arithmetic
print(a + b, a - b, a * b, a / b, a % b, a ** b)

# Comparison
print(a > b, a == b, a != b)
```

```python
# Input from user and print functions
user_name = input("Enter your name: ")
favourite_color = input("Enter your favourite color: ")
print("Hello,", user_name, "Your favourite color is", favourite_color)
```

```python
# Type Conversions
# Use the below function to convert a string into int, float and bool respectively
# int()
# float()
# bool()
birth_year = input("What is your birth year: ")
age = 2025 - int(birth_year)
print("Your age is", age)
```

```python
# Swapping of numbers without using third variable.
x = 5
y = 10
x = x + y
y = x - y
x = x - y
print(x,y)

# Printing all even numbers from 1 tp 20
for i in range(11):
    print(i*2, end = " ")
print("")

# Constructing dictionary with letters as keys and their AASCII as values
d = {"a": ord('a'),"b":ord('b'), "c":ord('c'), "d":ord('d'), "e":ord('e')}
print(d)
```
