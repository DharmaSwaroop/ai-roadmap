```python
# Functions using def
def hello_world():
    print("Hello World")

def hello_world2():
    return 'Hello_world'

print(hello_world) # Prints the address of that function
print(hello_world()) # Calls the function and prints whatever the function returns. In this case the function is returning none.
print(hello_world2()) # Same as above. But in this case the function is returning string.
print(hello_world2().upper()) # Prints all in upper case.
hello_world()
```

```python
# Functions with arguments and return_values
def hello_world(arg1, name='Dharma'):
    return f'{arg1}, {name}'

def student_info(*args, **kwargs):
    print(args)
    print(kwargs)

print(hello_world('hello', name="Swaroop")) # Ovverides the default key-word argument.
student_info('hello', 'Dharma', name='Swaroop', name2='Kaperla')

```

```python
# local and global values

Variable1 = 'String1' 
x = 6

def fun1():
    global x
    
    Variable1 = 'String2'
    print(x)
    x = x+5 # This wont work unless you explcitly mention that x is a global variable.
    print(Variable1)

def fun2():
    print(Variable1)
    print(x)

fun1()
fun2()
```

```python
# Lambda functions
print((lambda x,y: x+y)(3,5)) # small, anonymous (nameless) function defined using the keyword lambda

def add(x,y):
    return x+y
print(add(3,4))
```
